```python
df['Cpu Name']=df['Cpu'].str.split().str[:3].str.join(' ')
```

```python
import re
pattern = r'(\d+\.?\d*)\s*(?:TB|GB)\s*(SSD|HDD|Flash Storage|Hybrid)'

# Extract features using regular expressions
features = df['Memory'].str.extract(pattern, flags=re.IGNORECASE)

# Rename columns for clarity
features.columns = ['Storage Capacity', 'Storage Type']

# Concatenate 'features' DataFrame with the original DataFrame 'df'
df = pd.concat([df, features], axis=1)
```

```python
df.head()
```

```python
storage_type_dummies = pd.get_dummies(df['Storage Type'])

# Concatenate the binary columns with the original DataFrame 'df'
df = pd.concat([df, storage_type_dummies], axis=1)
```
